```python
#Question 3 function to have the factorial of a positive no
def pos_fact(n):
    fac = 1
    while n > 1:

        fac = fac * n
        n = n - 1
    return fac
# now testing the factorial function......
a = int(input())
print(pos_fact(a))
```

```python
#Ques4 function to print all the prime no in an interval.

def is_prime(n):
    for i in range(2,n,1):
        if n % i == 0:
            return False
    return True



def print_primes(s,l):
    count = 0
    for i in range(s,l+1,1):
        if is_prime(i) == True:
            count += 1
    return count

s = int(input())
l = int(input())

print(print_primes(s,l))

```

```python
#5.Getting the lcm of two user input nos
import math as m
def lcm(a,b):
    return a*b // m.gcd(a,b)
a = int(input())
b = int(input())
print(lcm(a,b))
```

```python
#14.scatter plot using points of ecllipse.
mu_x = 5
mu_y = -5
a = 5
b = 2.5
n_points = 100
points = []
while len(points) < n_points:
    x = np.random.uniform(mu_x-a,mu_x+a)
    y = np.random.uniform(mu_y-a,mu_y+a)
    if((x-mu_x)**2/a**2) + ((y-mu_y)**2/b**2) <= 1:
        points.append([x,y])
points = np.array(points)
x_points = points[:,0]
y_points = points[:,1]
df = pd.DataFrame({'x':x_points,'y':y_points})
print(df)
#scatter plot of points genrated threw ecllipse.
sns.scatterplot(x='x',y='y',data=df,label='Scatter plot genrated threw ecllipse') 
sns.kdeplot(x='x',y='y',data=df,label='kde plot')
```
