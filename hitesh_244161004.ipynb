```python
#Question 3 function to have the factorial of a positive no
def pos_fact(n):
    fac = 1
    while n > 1:

        fac = fac * n
        n = n - 1
    return fac
# now testing the factorial function......
a = int(input())
print(pos_fact(a))
```

```python
#Ques4 function to print all the prime no in an interval.

def is_prime(n):
    for i in range(2,n,1):
        if n % i == 0:
            return False
    return True



def print_primes(s,l):
    count = 0
    for i in range(s,l+1,1):
        if is_prime(i) == True:
            count += 1
    return count

s = int(input())
l = int(input())

print(print_primes(s,l))

```

```python
#Ques14.scatter plot using points of ecllipse.
mu_x = 5
mu_y = -5
a = 5
b = 2.5
n_points = 100
points = []
while len(points) < n_points:
    x = np.random.uniform(mu_x-a,mu_x+a)
    y = np.random.uniform(mu_y-a,mu_y+a)
    if((x-mu_x)**2/a**2) + ((y-mu_y)**2/b**2) <= 1:
        points.append([x,y])
points = np.array(points)
x_points = points[:,0]
y_points = points[:,1]
df = pd.DataFrame({'x':x_points,'y':y_points})
print(df)
#scatter plot of points genrated threw ecllipse.
sns.scatterplot(x='x',y='y',data=df,label='Scatter plot genrated threw ecllipse') 
sns.kdeplot(x='x',y='y',data=df,label='kde plot')
```

```python
#Question11.Reading the csv file as a numpy array
X = np.loadtxt("rdata.csv",delimiter=" ",dtype=float)

mu = np.mean(X, axis=0)  


C = np.zeros((10, 10))

for i in range(X.shape[0]):
    x_i = X[i, :]
    diff = x_i - mu
    C += np.outer(diff, diff)  

n = X.shape[0]
C /= n
print("Covariance matrix C:")
print(C)
```

```python
# Question 6.Bubble sort algorithm implementation
def bubble_sort(arr):
    n = len(arr)
    # Create a list to store the indices after sorting
    indices = list(range(n))
    
    for i in range(n):
        # Flag to check if any swap occurs
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] < arr[j+1]:  
                arr[j], arr[j+1] = arr[j+1], arr[j]
                # Swap the indices
                indices[j], indices[j+1] = indices[j+1], indices[j]
                swapped = True
        if not swapped:
            break  
    
    return arr, indices

# Create a list of length n=15 with random integers
import random
n = 15
random_list = [random.randint(1, 100) for _ in range(n)]

# Perform bubble sort on the list and get the sorted list and indices
sorted_list, sorted_indices = bubble_sort(random_list)

# Print the sorted list and the indices
print("Sorted list:", sorted_list)
print("Sorted indices:", sorted_indices)
```

```python
#Question 7.

def bubble_sort(arr):
    n = len(arr)
   
    indices = np.arange(n)
    
    for i in range(n):
        
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]: 
              
                arr[j], arr[j+1] = arr[j+1], arr[j]
              
                indices[j], indices[j+1] = indices[j+1], indices[j]
                swapped = True
        if not swapped:
            break  
    
    return arr, indices


n = 15
random_array = np.random.randint(1, 100, size=n)

sorted_array, sorted_indices = bubble_sort(random_array)

print("Sorted array:", sorted_array)
print("Sorted indices:", sorted_indices)
```

```python

#Question 9.Matrix multiplication with and without built in functions.
m, n, p = 3, 4, 2  


M1 = np.random.rand(m, n)
M2 = np.random.rand(n, p)


def matrix_multiply_manual(M1, M2):
    m, n = M1.shape
    n2, p = M2.shape
    if n != n2:
        raise ValueError("The number of columns in M1 must match the number of rows in M2.")
    
   
    result = np.zeros((m, p))
    
    
    for i in range(m):  
        for j in range(p):  
            for k in range(n): 
                result[i, j] += M1[i, k] * M2[k, j]
    
    return result


result_numpy = np.dot(M1, M2)


result_manual = matrix_multiply_manual(M1, M2)


print("Matrix M1:")
print(M1)
print("\nMatrix M2:")
print(M2)
print("\nResult using NumPy's dot function:")
print(result_numpy)
print("\nResult from manual multiplication:")
print(result_manual)

print("\nDo the results match? ", np.allclose(result_numpy, result_manual))
```

```python
#Question 12.conjugate,magnitude,addition,subtraction,multiplicaton,division,string reprenstation and angle using object oriented programming.
import cmath
class ComplexNumber:
    def __init__(self,real,imag):
        self.real = real
        self.imag = imag
    def __str__(self):
        return f"{self.real} + {self.imag}j"
    def conjugate(self):
        return ComplexNumber(self.real,-self.imag)
    def absolute_value(self):
        return (self.real**2 + self.imag**2)**0.5
    def add(self, other):
        return ComplexNumber(self.real + other.real, self.imag + other.imag)
    def subtract(self, other):
        return ComplexNumber(self.real - other.real, self.imag - other.imag)
    def multiply(self, other):
        real = self.real * other.real - self.imag * other.imag
        imag = self.real * other.imag + self.imag * other.real
        return ComplexNumber(real, imag)
    def divide(self, other):
        denominator = other.real**2 + other.imag**2
        real = (self.real * other.real + self.imag * other.imag) / denominator
        imag = (self.imag * other.real - self.real * other.imag) / denominator
        return ComplexNumber(real, imag)
    def angle(self):
        return cmath.phase(complex(self.real, self.imag))
# Define two complex numbers
c1 = ComplexNumber(100,110)  # Complex number 3 + 4j
c2 = ComplexNumber(17, 22)  # Complex number 1 + 2j
# Perform operations
print(f"c1 = {c1}")
print(f"c2 = {c2}")
# Addition
print(f"c1 + c2 = {c1.add(c2)}")
# Subtraction
print(f"c1 - c2 = {c1.subtract(c2)}")
# Multiplication
print(f"c1 * c2 = {c1.multiply(c2)}")
# Division
print(f"c1 / c2 = {c1.divide(c2)}")
# Absolute value
print(f"|c1| = {c1.absolute_value()}")
print(f"|c2| = {c2.absolute_value()}")
# Angle (phase)
print(f"c1 = {c1.angle()} radians")
print(f"c2 = {c2.angle()} radians")
# Conjugate
print(f"Conjugate of c1 = {c1.conjugate()}")
print(f"Conjugate of c2 = {c2.conjugate()}")
```
